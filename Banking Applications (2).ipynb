```python
balance=0
prev_trans=[]

def check_balance():
    global prev_trans
    global balance
    prev_trans=check_balance
    print("Your Balance is",balance)
    print("-------------------------------------------------------------------------------------------------")
    
def deposit():
   global balance
   global prev_trans
   deposit_amount=int(input("Enter the Amount to Deposit:"))
   balance=balance+deposit_amount
   prev_trans = ["deposit", deposit_amount]
   print(balance,"Credited to your Account")
   print("-------------------------------------------------------------------------------------------------")
   
def withdraw():
    global balance
    if withdraw_amount>balance:
        withdraw_amount=int(input("Enter the amount to withdraw:"))
        balance=balance-withdraw_amount
        prev_trans=["withdraw",withdraw_amount]
        print(balance,"Debited from your Account")
        print("-------------------------------------------------------------------------------------------------")
    else:
        print("No Balance is Available ")
        print("-------------------------------------------------------------------------------------------------")
    
def previous_transaction():
    global prev_trans
    try:
        if not prev_trans:
            print("No transactions have occurred yet.")
            print("-------------------------------------------------------------------------------------------------")
        else:
            trans_type,amount=prev_trans
            if trans_type == 'deposit':
                print("Previous transaction: Deposit", amount)
                print("-------------------------------------------------------------------------------------------------")
            elif trans_type == 'withdraw':
                print("Previous transaction: Withdraw", amount)
                print("-------------------------------------------------------------------------------------------------")
    except:
        print("No transactions have occurred yet")
def exit():
    print("You have exited")
    print("Thank you for choosing our Bank")
    print("-------------------------------------------------------------------------------------------------")

def user_choosing():
    while True:
        print("1. Check Balance")
        print("2. Deposit Amount")
        print("3. Withdraw Amount")
        print("4. Previous Transaction")
        print("5. Exit")
        options_input = int(input("What do you want to do?"))

        if options_input == 1:
            check_balance()
        elif options_input == 2:
            deposit()
        elif options_input == 3:
            withdraw()
        elif options_input == 4:
            previous_transaction()
        elif options_input == 5:
            exit()
            break
        else:
            print("Enter a valid option")

while True:
    customer_name = input("Enter Your Name:").strip()
    if customer_name.isalpha():
        break
    else:
        print("Invalid name. Please enter only alphabets.")
while True:
    try:
        account_number = int(input("Enter Your Account Number:"))
        break
    except ValueError:
        print("Invalid Account Number. Please enter a valid numeric account number.")
print("-------------------------------------------------------------------------------------------------")
print("Welcome", customer_name)
print("Your Account Number is", account_number)
user_choosing()

```
